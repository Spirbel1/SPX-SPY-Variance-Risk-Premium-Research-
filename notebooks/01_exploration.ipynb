# SPX/SPY VRP Exploration
Load engineered dataset and inspect core VRP relationships.

```python
from pathlib import Path
import pandas as pd
import plotly.express as px

root = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
dataset_path = root / 'data' / 'processed' / 'vrp_dataset.parquet'
df = pd.read_parquet(dataset_path)
df = df.reset_index().rename(columns={'index': 'date'}) if 'date' not in df.columns else df
df['date'] = pd.to_datetime(df['date'])
df[['date','close','vix','VRP_21d','next_5d_return']].dropna().tail()
```

```python
plot_df = df[['date', 'VRP_21d', 'next_5d_return']].dropna()
fig = px.scatter(plot_df, x='VRP_21d', y='next_5d_return', trendline='ols', title='VRP vs Next 5d Return')
fig.show()
```
